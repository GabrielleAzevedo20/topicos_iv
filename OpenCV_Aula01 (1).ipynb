```python
# Exemplo de uso de conversão de cores. Convertendo uma imagem de BGR para escala de cinza:
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# Exemplo de uso de conversão de cores. Convertendo uma imagem de BGR para escala de RGB:
imgBGR = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(imgBGR)
```

```python
# Função de desenho: desenhando círculo + retângulo:
cv2.circle(img, (200, 100), 85, (100,255,0), 3 )
cv2_imshow(img)
```

```python
from google.colab.patches import cv2_imshow
import cv2

cap = cv2.VideoCapture('video.mp4')
while cap.isOpened():
    ret, image = cap.read()

    if not ret:
      break

    cv2_imshow(image) # Note cv2_imshow em vez de cv2.imshow

    cv2.waitKey(1) & 0xff

cv2.destroyAllWindows()
cap.release()
```
